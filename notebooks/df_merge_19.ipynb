```python
citibikedemand['key'] = citibikedemand['month'].astype(str) +\
                        citibikedemand['day'].astype(str) +\
                        citibikedemand['night?'].astype(str) +\
                        citibikedemand['hex_id']
```

```python
citibikedemand.head()
```

```python
citibikedemand = citibikedemand[['citi_inflow_counts','citi_outflow_counts','key']]
```

```python
merged['year'] = 2019
```

```python
merged.head()
```

```python
mtademand = pd.read_csv('../Capstone_Project/cusp_capstone/data_for_notebooks/clena_df19_new.csv')
```

```python
mtademand.head(10)
```

```python
mtademand['month']  = mtademand.apply(lambda row: row['date'].split('/')[0], axis=1)
mtademand['day']    = mtademand.apply(lambda row: row['date'].split('/')[1], axis=1)
```

```python
mtademand['month']  = mtademand['month'].astype(int)
mtademand['day']    = mtademand['day'].astype(int)
```

```python
mta19 = mtademand[mtademand['month']<5]
```

```python
hourgroups = {'8pm - 12am':1, 
              '12am - 4am':1, 
              '4am - 8am':0,
              '8am - 12pm':0,
              '12pm - 4pm':0,
              '4pm - 8pm':0}
```

```python
mta19['night?'] = mta19['timegroupstr'].map(hourgroups)
mta19['night?'] = mta19['night?'].astype(int)
```

```python
mta19.head(10)
```

```python
MTA_hex = pd.read_csv("../Capstone_Project/cusp_capstone/data_for_notebooks/mta_cleaned.csv")
```

```python
MTA_hex.shape
```

```python
mta19_hex = pd.merge(mta19, MTA_hex, how='inner',left_on='c/a',right_on='Booth')
mta19_hex.head()
```

```python
# pivot table to sum day/night rides by hexagon, day, and month
mta19_cleaned = pd.pivot_table(mta19_hex,
                       values=['entries_diff', 'exit_diff'],
                       index=['hex_id_8', 'month', 'day', 'weekday', 'night?'],
                       aggfunc=np.sum).reset_index()
```

```python
mta19_cleaned = mta19_cleaned.rename(columns={"entries_diff": "mta_inflow_counts", "exit_diff": "mta_outflow_counts"})
```

```python
mta19_cleaned['key'] = mta19_cleaned['month'].astype(str) +\
                        mta19_cleaned['day'].astype(str) +\
                        mta19_cleaned['night?'].astype(str) +\
                        mta19_cleaned['hex_id_8']
```

```python
mtademand = mta19_cleaned[['mta_inflow_counts','mta_outflow_counts','key']]
```

```python
merged19 = merged.merge(mtademand, how ='left', on='key')
```

```python
merged19 = merged19.fillna(0)
```

```python
merged19.tail()
```

```python
merged19.to_file("merged19.geojson",
                         driver='GeoJSON')
```

```python
weather2019 = pd.read_csv('../Capstone_Project/cusp_capstone/data_for_notebooks/2019_Weather.csv',index_col=0)
```

```python
weather2019['month']  = weather2019.apply(lambda row: row['just_date'].split('/')[0], axis=1)
weather2019['day']    = weather2019.apply(lambda row: row['just_date'].split('/')[1], axis=1)

weather2019['month']  = weather2019['month'].astype(int)
weather2019['day']    = weather2019['day'].astype(int)
```
